## **Зад 1. Проверка за вътрешно произведение**

- Ковариация. 

````python
import numpy as np

# Данни за тестване
x_test = np.array([1, 2, 3, 5])
y_test = np.array([1, 2, 4, 6])

def is_symmetric(f):
    for x in x_test:
        for y in y_test:
            if f(x, y) != f(y, x):
                return False
    return True

def is_positive_definite(f):
    for x in x_test:
        if f(x, x) < 0:
            return False
    return True

# B(u + v,w) = B(u,w) + B(v,w) and B(u,v + w) = B(u,v) + B(u,w)
def check_additivity(f):
    for u in x_test:
        for v in x_test:
            for w in y_test:
                if f(u + v, w) != f(u, w) + f(v, w):
                    return False
                if f(u, v + w) != f(u, v) + f(u, w):
                    return False
    return True

# B(λu,w) = λB(u,w) and B(u,λv) = λB(u,v)
w_test = np.array([1, 2, 6, 7])
def check_homogeneity(f, lam):
    for u in x_test:
        for v in y_test:
            for w in w_test:
                if f(lam * u, w) != lam * f(u, w):
                    return False
                if f(u, lam * v) != lam * f(u, v):
                    return False
    return True

def is_bilinear(f, lam=2):
    return check_additivity(f) and check_homogeneity(f, lam)

def f(x, y):
    # return np.dot(x, y)
    cov_matrix = np.cov(x, y, ddof=0)
    return cov_matrix[0, 1]

if is_symmetric(f) and is_positive_definite(f) and is_bilinear(f):
    print("f е вътрешно произведение")
else:
    if not is_symmetric(f):
        print("f не е симетрична")
    if not is_positive_definite(f):
        print("f не е положително дефинирана")
    if not is_bilinear(f):
        print("f не е билинейна")
````

## **Зад 2. Теорема на Питагор**

````python
import numpy as np

x = np.array([1, -1, 0, 0])
y = np.array([0, 0, 1, -1])

# Центриране премахва средната стойност от данните, което гарантира, 
# че ковариацията и вътрешното произведение измерват само линейните зависимости между отклоненията.
x_centered = x - np.mean(x)
y_centered = y - np.mean(y)

cov_matrix = np.cov(x_centered, y_centered, ddof=0)
inner_product = cov_matrix[0, 1]

norm_x = np.sqrt(cov_matrix[0,0])
norm_y = np.sqrt(cov_matrix[1,1])

cos_theta = inner_product / (norm_x * norm_y)
print("cos(theta) =", cos_theta)

def pythagoras_theorem(norm_x, norm_y, inner_product):
    if np.isclose(inner_product, 0): 
        norm_sum = np.sqrt(norm_x**2 + norm_y**2)
        return norm_sum
    else:
        return None  

if (cos_theta == 0):
    sigma_sum = pythagoras_theorem(norm_x, norm_y, inner_product)
    print(sigma_sum)
    print("Питагоровата теорема е в сила.")
else:
    print("Питагоровата теорема не е в сила.")
````
