# **Домашна №5**
## Прочетете **5-4-new-sssp.pdf**
Статията `Breaking the Sorting Barrier for Directed Single-Source Shortest Paths (2025)` представя нов детерминиран алгоритъм с времева сложност `O(m log²⁄₃ n)` за намиране на най-кратки пътища от един източник в насочени графи с реални ненегативни тегла. Това е първото доказателство, че е възможно да се преодолее класическата граница на сортиране, наложена от алгоритъма на `Dijkstra`, който има времева сложност `O(m + n log n)` в разреждени графи.
