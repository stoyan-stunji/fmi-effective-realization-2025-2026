````python
import numpy as np
import matplotlib.pyplot as plt

A = 2 * np.eye(2)

# A = np.array([[-1, 0],
#               [0, -1]])

# k = 1.0  # horizontal shear
# s = 1.5  # vertical scaling

# H = np.array([[1, k],
#              [0, 1]])
# S = np.array([[1, 0],
#              [0, s]])
# A = np.dot(S, H)

angles = np.linspace(0, 2 * np.pi, 10)
vectors = np.array([ [np.cos(a), np.sin(a)] for a in angles ])

scaled_vectors = np.dot(vectors, np.transpose(A))

plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)

for v in vectors:
    plt.arrow(0, 0, v[0], v[1], color='blue', alpha=0.5, head_width=0.05)

for v in scaled_vectors:
    plt.arrow(0, 0, v[0], v[1], color='red', alpha=0.8, head_width=0.05)

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.gca().set_aspect('equal')
plt.title("Blue: original, Red: scaled х2")
plt.show()
````

## **Зад 4. Power Iteration (Von Mises Iteration)**
- Алгоритъм за собствени стойности. Нека имаме матрица `А`, която може да се диагонализира. Алгоритъмът ще върне стойност `a`, която е най-голямата собствена стойност на `А`, и ненулев вектор `b`, съответстващ на собствената стойност `a`, такъв че `Ab` = `ab`. Алгоритъмът започва със случаен вектор `b_0`. Методът се описва от рекурентното уравнение: 

$$
b_{k+1} = \frac{A b_k}{\| A b_k \|}
$$

- Намерете грешката в кода? Защо няма да работи? Коя операция отнема най-много време? За какъв тип матрици ще работи най-ефективно?

``` python
import numpy as np

def power_iteration(A, num_iterations: int):
    b_k = np.random.rand(A.shape[1])
    for _ in range(num_iterations):
        b_k1 = np.dot(b_k, A)
        b_k = b_k1 / np.linalg.norm(b_k1)
    return b_k
    
power_iteration(np.array([[0.5, 0.5], [0.2, 0.8]]), 10)
```

$$
A = \begin{bmatrix} 
A_{11} & A_{12} \\ 
A_{21} & A_{22} 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
B_{11} & B_{12} \\ 
B_{21} & B_{22} 
\end{bmatrix}
$$

$$
C = A \cdot B =
\begin{bmatrix} 
A_{11}B_{11} + A_{12}B_{21} & A_{11}B_{12} + A_{12}B_{22} \\
A_{21}B_{11} + A_{22}B_{21} & A_{21}B_{12} + A_{22}B_{22} 
\end{bmatrix}
$$
