### **Зад 4. Kruskal's Algorithm**
- $Kruskal$ намира $MST$ в свързан, неориентиран граф с тегла на ребрата. Алгоритъмът постепенно добавя най-евтините ръбове, като се избягват цикли, докато всички върхове са свързани в минимално остовно дърво. За проверка на цикли се използва структурата $Union$-$Find$.

- Алгоритъм:
    - Сортирай всички ребра по нарастващо тегло;
    - Инициализирай $MST$ като празно множество;
    - Инициализирай $Union$-$Find$ структура за всички върхове;
    - Докато $MST$ не съдържа $V-1$ ребра:
        - Вземи реброто с минимално тегло $(u, v)$;
        - Ако $u$ и $v$ принадлежат на различни множества (не създават цикъл):
            - Добави реброто към $MST$;
            - Обедини множествата на $u$ и $v$ в Union-Find;
        - Повтори, докато $MST$ съдържа всички върхове.

- $Union$-$Find$:
    - $Find(x)$: намира представителя на множеството, към което принадлежи $x$;
    - $Union(x, y)$: обединява множествата на $x$ и $y$;

### **Зад 5. Dijkstra's Algorithm**
- $Dijkstra$ намира най-кратките пътища от даден начален връх до всички останали върхове в тегловен граф без отрицателни тегла на ребрата.
- Алгоритъм:
    - Инициализация:
        - За всеки връх $v$ в графа задай $dist[v] = \infty$;
        - Задай $dist[start] = 0$;
        - Добави всички върхове в приоритетна опашка според тяхната текуща стойност на $dist$;

        - Извади върха $u$ с минимална стойност на $dist[u]$;

                - Обнови $dist[v] = dist[u] + w$;
                - Обнови позицията на $v$ в приоритетната опашка.

### **Зад 6. Uniform Cost Search**
- $UCS$ намира най-краткия път от начален връх до даден целеви връх в граф с неотрицателни тегла на ребрата.
- Алгоритъм:
    - Инициализация:
        - За всеки връх $v$ в графа задаваме $cost[v] = \infty$;
        - Задаваме $cost[start] = 0$;
        - Добавяме началния връх в приоритетна опашка според текущата стойност на $cost$;
    - Докато приоритетната опашка не е празна:
        - Изваждаме върха $u$ с най-малка стойност на $cost[u]$;
        - Ако $u$ е целевият връх, спираме и връщаме $cost[u]$ като най-кратък път;
        - За всеки съсед $v$ на $u$ с тегло $w$:
            - Ако $cost[u] + w < cost[v]$, тогава:
                - Обновяваме $cost[v] = cost[u] + w$;
                - Добавяме $v$ в приоритетната опашка с новия $cost[v]$.

- Сравнение с $Dijkstra$:
    - $Dijkstra$ намира най-кратките пътища до всички върхове, докато $UCS$ може да спре при достигане на целевия връх;
    - И двата използват приоритетна опашка и работят с неотрицателни тегла;
    - $UCS$ се разглежда като специален случай на $Dijkstra$, фокусиран върху намиране на най-кратък път до конкретна цел.

### **DFS алгоритъм за проверка на двуделен граф**
- Инициализация:
  - За всеки връх $v$ задаваме $color[v] = \text{None}$;
- За всеки непосетен връх $v$:
  - Ако $DFS(v, цвят 0)$ върне $False$, графът не е двуделен;
- $DFS(u, текущ_цвят)$:
  - Задаваме $color[u] = текущ\_цвят$;
    - За всеки съсед $v$ на $u$:
      - Ако $color[v] = \text{None}$:
        - Ако $DFS(v, 1 - текущ\_цвят)$ върне $False$, връщаме $False$;
      - Иначе ако $color[v] = текущ\_цвят$:
        - Връщаме $False$ (открит е нечетен цикъл);
      - Връщаме $True$.
- Ако $DFS$ завърши успешно за всички върхове без конфликти, графът е двуделен.

### **Зад 10. Yen's Algorithm**

- Какво прави следният код?

``` java
package Algorithms;

import Graph.Edge;
import Graph.Graph;
import Graph.Path;
import Graph.Vertex;

import java.util.*;
import java.util.stream.Collectors;

public class Yen {
    private final Graph graph;
    private final Dijkstra dijkstra;

    public Yen(Graph graph) {
        this.graph = graph;
        this.dijkstra = new Dijkstra(graph);
    }

    public List<Path> calculate(Vertex source, Vertex target, int k) {
        List<Path> paths = new ArrayList<>();
        Path shortestPath = dijkstra.getShortestPath(source, target, null);
        if (shortestPath == null) {
            return paths;
        }

        paths.add(shortestPath);
        Set<Edge> exclusionCandidates = new HashSet<>(shortestPath.getEdges(graph));

        for (int i = 1; i < k; i++) {
            Path minPath = findNextShortestPath(paths.get(i - 1), source, target, exclusionCandidates);
            if (minPath == null) {
                break;
            }
            paths.add(minPath);
            exclusionCandidates.addAll(minPath.getEdges(graph));
        }
        return paths;
    }

    private Path findNextShortestPath(Path previousPath, Vertex source, Vertex target, Set<Edge> exclusionCandidates) {
        Path minPath = null;
        List<Vertex> vertices = previousPath.getVertices();

        for (int j = 0; j < vertices.size() - 1; j++) {
            Vertex root = vertices.get(j);
            Path spurPath = buildSpurPath(source, root, target, exclusionCandidates);
            if (spurPath == null || !spurPath.getVertices().contains(target)) {
                continue;
            }
            if (minPath == null || spurPath.getWeight() < minPath.getWeight()) {
                minPath = spurPath;
            }
        }
        return minPath;
    }

    private Path buildSpurPath(Vertex source, Vertex root, Vertex target, Set<Edge> exclusionCandidates) {
        Path rootPath = dijkstra.getShortestPath(source, root, null);
        Set<Edge> excludedEdgesFromRoot = exclusionCandidates.stream()
                .filter(edge -> edge.source().equals(root))
                .collect(Collectors.toSet());

        Path spur = dijkstra.getShortestPath(root, target, excludedEdgesFromRoot);
        if (spur == null) {
            return null;
        }
        return rootPath.join(spur, true);
    }
}
```

### **Bellman-Ford's Algorithm**

- Релаксира всички ребра многократно - точно $V − 1$ пъти, където $V$ е броят на върховете. След тези итерации разстоянията са гарантирано най-къси.
- Интуиция:
    - Най-късият път между два върха може да има най-много $V − 1$ ребра;
    - Затова, ако подобряване на разстоянията $V − 1$ пъти, всички възможни пътища вече ще са оптимизирани.
- Алгоритъм:
    - За всеки връх $v$:
           - $dist[v] = ∞$
       - $dist[s] = 0$
    - Повтори $V − 1$ пъти:
       - за всяко ребро $(u, v)$ с тегло $w$:
           - ако $dist[u] + w < dist[v]$:
               - $dist[v] = dist[u] + w$
    - Проверка за отрицателни цикли:
       - за всяко ребро $(u, v)$ с тегло $w$:
           - ако $dist[u] + w < dist[v]$:
               - има отрицателен цикъл

````python
import numpy as np

def bellman_ford(n, edges, source):
    dist = np.full(n, np.inf)
    dist[source] = 0.0

    for i in range(n - 1):
        updated = False
        for (u, v, w) in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                updated = True
        if not updated:
            break

    for (u, v, w) in edges:
        if dist[u] + w < dist[v]:
            raise ValueError("Графът има отрицателен цикъл!")

    return dist

edges = np.array([
    [0, 1, 4],
    [0, 2, 2],
    [2, 1, -1],
    [1, 3, 2],
    [2, 3, 3]
])

n = 4         
source = 0     

distances = bellman_ford(n, edges, source)
for v in range(n):
    print(f"dist[{v}] = {distances[v]}")
````
